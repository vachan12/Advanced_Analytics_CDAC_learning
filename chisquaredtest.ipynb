## CHI SQUARE TEST


### chi-square test of independence: test of two categorical columns
### H0: No association between passenger class and survival


```python
import  pandas as pd
from scipy.stats import chi2_contingency
df = pd.read_csv("titanic.csv")
df = df[['pclass','survived']].dropna()

#create contingency table
contingency_table = pd.crosstab(df['pclass'],df['survived'])
print(contingency_table)
#Perform chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

print("\n Chi-Square statistic: ",chi2_stat)
print("p-value: ",p_value)
print("\n Expected Frequencies: \n", expected)
alpha=0.05
if p_value >alpha:
    print("Fail to reject H0")
else:
    print("Reject H0")
```
