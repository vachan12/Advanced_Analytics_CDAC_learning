```python
import pandas as pd
from scipy.stats import zscore, norm

df = pd.read_csv('diabetes.csv')

glucose_mean = df['Glucose'].mean()
glucose_std = df['Glucose'].std()
print(f"Mean of Glucose: {glucose_mean}")
print(f"Standard Deviation of Glucose: {glucose_std}")

# Calculate z-score and probability for each row
df['Glucose_zscore'] = zscore(df['Glucose'])
df['Glucose_probability_cdf'] = norm.cdf(df['Glucose_zscore'])
print(df)

# Find the z-score at the 30th percentile
z_score_30 = norm.ppf(0.30)
print(f"Z-score at the 30th percentile: {z_score_30}")

# Get the probability (CDF) for this specific z-score
probability_30 = norm.cdf(z_score_30)
print(f"Probability (CDF) for z-score {z_score_30}: {probability_30:.4f}")
```

```python
df.head()
#shapiro test and QQ plot to check data normality
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import shapiro

df = pd.read_csv("diabetes.csv")
print(df.head())

#Plot a histogram of blood pressure column
plt.hist(df['BloodPressure'],bins = 30,edgecolor='k',alpha=0.7)
plt.title("Blood pressure Distribution")
plt.xlabel("Blood Pressure")
plt.ylabel("frequency")
plt.show()

#Generate a Q-Q plot
stats.probplot(df['BloodPressure'],dist='norm',plot= plt)
plt.title("QQ plot of blood pressure")
plt.show()

#Perform the Shapiro-Wilk test
shapiro_test = shapiro(df['BloodPressure'])
print("shapiro-Wilk p-values:",shapiro_test.pvalue)

#Interpret the result 
alpha = 0.5

#HQ: DataFrame seems to be normally distributed
if shapiro_test.pvalue > alpha:
    print("Fail to reject H0")
else:
    print("Reject H0")
print('skewness:',df['BloodPressure'].skew())
```

## Central Limit Theorem

```python
#CLT - Titanic

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import shapiro

df=pd.read_csv("titanic.csv")

#removing missing ages
ages = df['age'].dropna()
print(df.head())

#Shapiro test on Age column

shapiro_test = stats.shapiro(ages)
print("Shapiro-Wilk p-value:",shapiro_test.pvalue)

if shapiro_test.pvalue < 0.05:
    print("The data likely does not follow a normal distribution.")
else:
    print("The data may be normally distributed, but the q-q plot can provide further insights.")

#Generate a Q-Q plot for Age
stats.probplot(ages,dist="norm",plot=plt)
plt.title("Q-Q Plot for Age")
plt.show()

#Central Limit Theorem
sample_means = []
n_samples = 100
sample_size = 30

for _ in range(n_samples):
    sample= ages.sample(n=sample_size,replace=True)
    sample_means.append(sample.mean())

#Plot the distribution sample means
plt.hist(sample_means,bins=30,edgecolor='k',alpha=0.7)
plt.title("Distribution of sample Means (n=30)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()

#Generate a Q-Q plot for the sample  means
stats.probplot(sample_means,dist="norm",plot=plt)
plt.show()

#Perform the Shapiro-wilk Test on the sample means
shapiro_test =  stats.shapiro(sample_means)
print("Shapiro-Wilk p-value", shapiro_test.pvalue)

#Interpret the result
if shapiro_test.pvalue < 0.05:
    print("The sample means likely do not follow a normal distribution.")
else:
    print("The sample means may be normally distributed, supporting the Central Limit Theo")
```
