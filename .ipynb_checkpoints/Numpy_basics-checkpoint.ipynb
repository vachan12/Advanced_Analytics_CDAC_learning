```python
import numpy as np
```

```python

print("==================creating array=====================")
arr = np.array([1,2,3,4])
lst = [1,2,3,4]
print("==================Add  5=====================")
print(arr+5)
print([x+5 for x in lst])
print("==================multiply by 2=====================")
print(arr*2)
print([x*2 for x in lst])
print("==================Squares=====================")
print(arr**2)
print([x**2 for x in lst])
print("==================Mean=====================")
print(np.mean(arr))
print(sum(lst)/len(lst))
print("==================Filtering factors of 2=====================")
print(arr[arr % 2 == 0])
print([x for x in lst if x%2==0])
print("==================2D Array=====================")
print(np.array([[1,2],[3,4]]))
print([[1,2],[3,4]])
print("==================Adding Two Arrays=====================")
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a+b)

a=[1,2,3]
b=[4,5,6]
print([x+y for x,y in zip(a,b)])

print("==================Broadcasting=====================")
'''Broadcasting is a mechanism that allows arrays of different shapes to perform arithmetic operations together without 
explicitly resizing or copying data'''
print(arr+1)
#python list: Not directly supported like NumPy

print("===================Condtional Replace===============")
#Conditional Replace
arr[arr>2] = 0
print(arr)

lst = [1,2,3,4]
print([0 if x>2 else x for x in lst])
```

```python
'''comparing arange() and linspace()'''
a=np.arange(0,10,2)
b=np.linspace(0,10,5)
print(len(a))
print(a)
print(len(b))
print(b)
```

```python
arr = np.arange(55,66)
print(arr)
```

```python
mask = arr > 58
print(mask)
```

```python

filtered = arr[mask]
print(filtered)
```

```python
#Multiple Conditions
print(arr[(arr>57) & (arr < 63)])
```

```python
#marks example
marks = np.array([67,34,89,32])
passed_boolean = marks > 35
print(passed_boolean)
print(marks[passed_boolean])
```

```python
#shallow v/s deep copy
a=np.array([10,20,30,40,50,60])

#shallow copy
b = a[0:5]
print(a)
print(b)

b[0] = 999
print(a)
print(b)

#Deep Copy
c=a.copy()

c[1] = 888
print(a)
print(c)

```
