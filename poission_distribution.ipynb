```python
#poisson Distribution
#A factory produces on an average 8 defective bulbs a day.
#What is the probability that is produces exactly 8 defective bulbs today?

from scipy.stats import poisson
mu=8 #average
x=8 #defective
print(f"probability that is produces exactly 8 defective bulbs today:{poisson.pmf(x,mu)}")
```

```python
#A life insuarance salesman sells on the average 3 life insuarance policies per week. Use Poission's law to calculate the probability 
#a.In a given week he will sell some policies
# b.In a given week he will sell two or more policies but not more than 4
# c.what is the probability that he will sell exactly 5 policies in a given week?
# d.what is the probability that he will sell atmost 1 policy in a given week?

mu=3
p=0
for i in range(1,4):
    p+=poisson.pmf(i,mu)
print("a)")
print(f"In a given week he will sell some policies: {p}")
p=0
for i in range(2,5):
    p+=poisson.pmf(i,mu)
print("b)")
print(f"In a given week he will sell two or more policies but not more than 4: {p}")

x=5
p=poisson.pmf(x,mu)
print("c)")
print(f"the probability that he will sell exactly 5 policies in a given week:{p}")

print("d)")
prob=0
for i in range(2):
    prob+=poisson.pmf(i,mu)
print(f"what is the probability that he will sell atmost 1 policy in a given week:{prob}")
```

```python
#A popular blog receives an average of 12 clicks per minute on a certain page. Use poisson's law to calculate the probability:
# a. In a given minute, the page will receive no clicks.
# b. In a given minute, the page will receive more than 10 clicks.
# c. In a given minute, the page will receive exactly 15 clicks.
# d. In a given minute, the page will receive atmost 13 clicks.
```

```python
# a. In a given minute, the page will receive no clicks.
mu=12
x=0
d=poisson.pmf(x,mu)
print(f"Probability that for a given minute, the page will receive no clicks is: {d:.12f}")
```

```python
# b. In a given minute, the page will receive more than 10 clicks.
prob=0
for i in range(11,13):
    prob+=poisson.pmf(i,mu)
print(f"Probability that for a given minute, the page will receive more than 10 clicks is: {prob}")
```

```python
# c. In a given minute, the page will receive exactly 15 clicks.
mu=12
x=15
d=poisson.pmf(x,mu)
print(f"Probability that for a given minute, the page will receive exactly 15 clicks is: {d:.12f}")
```

```python
# d. In a given minute, the page will receive atmost 13 clicks.
prob=0
for i in range(0,14):
    prob+=poisson.pmf(i,mu)
print(f"Probability that for a given minute, the page will atmost 13 clicks is: {prob}")
```
