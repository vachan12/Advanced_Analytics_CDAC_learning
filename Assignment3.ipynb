```python
# Q3.Two-sample t test - Assignment 3
# 1.Titanic: Is the average age of male passengers significantly different from the average age of female passengers?

# 2.Titanic: Is there a significant difference in age between survivors and non survivors?

# 3.diabetes: Is the mean glucose level significantly different between diabetic and non-diabetic patients ?

# 4.Diabetes : Is there a significant difference in average BMI between diabetic and non-diabetic patients ?

# 5.Tips : Do weekend customers spend more than week day customer?

import pandas as pd
from scipy import stats

df = pd.read_csv('titanic.csv')
df=df.dropna()

df['age'] =df['age']
male = df[df['gender']=='male']
male_mean = male['age'].mean()
print(f"male mean: {male_mean}")

female = df[df['gender']=='female']
female_mean = female['age'].mean()
print(f"female mean: {female_mean}")

male_sample = male['age'].sample(n=30,replace=True)
#print(male_sample)
female_sample = female['age'].sample(n=30, replace = True)
#print(female_sample)


# #Perform two-sample t-test

t_statistic, p_value = stats.ttest_ind(male_sample, female_sample)

# #Degrees of freedom
n_male = len(male_sample)
n_female = len(female_sample)
degrees_of_freedom = n_male + n_female-2
#Significance level (Alpha)
alpha = 0.05
#Find the critical value
critical_value = stats.t.ppf(1-alpha,degrees_of_freedom)

#Print the results
print("T-statistic: ", t_statistic)
print("P-value: ", p_value)

if p_value < alpha:
    print("There is a significant difference")

else:
    print("There is no significant difference")

```

```python
#Q6. Do multiple linear regression on insurance.csv dataset using X= age and bmi na dy= charges. Do not have visualisation code, since we have 2 X columns
import pandas as pd



#step1: Separate data into X and Y
df = pd.read_csv("insurance.csv")
X= df[['age','bmi']]
y=df['charges']

#Step2: Split the data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)

#Print X and y for the test set
print("---------Test set Data------------")
print("X test (age,bmi): \n", X_test)
print("\ny_test (charges):\n", y_test)
print("----------------\n")

#step3: Fit Simple Linear Regression to Training Data 
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)

#Step 4: make Prediction 
y_pred = model.predict(X_test)
print("Predict values on test set: ", y_pred)

#Step 6: Make new prediction
new_X_values_df = pd.DataFrame({'age': [1,3,5,8,12,15,20,25],'bmi':[78,55,77,8,545,55,42,0]})

new_y_pred = model.predict(new_X_values_df)
print('\nPredictions for new experience:\n', new_y_pred)

#Step 7 - Model Evaluation

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\nModel Evaluation Metrics:")
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error(MSE):", mse)
print("R2 score:", r2)
```
