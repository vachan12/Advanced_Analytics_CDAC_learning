## 1-sample T-test to check if averagae tip paid by a customer is close to $3

#H0: The avaerage tip paid by a customer is not different from $3

```python
import pandas as pd
from scipy import stats

df = pd.read_csv('tips.csv')
our_column=df['tip']
our_column = our_column.sample(n=30,replace = False)

meu= 3 #hypothesized population mean = $3

#PErform one-Sample t-test, 
t_statistic , p_value = stats.ttest_1samp(our_column, meu)

print("p-value:",p_value)
print(f"test statistic: {t_statistic}")

alpha = 0.05

if p_value < alpha:
    print("Reject H0")
else:
    print("Do not reject H0")
```

#1-Sampled T-test on marathon completion Time

```python
import pandas as pd
from scipy import stats
#LOAD DATASETS
df_2015 = pd.read_csv('marathon_results_2015.csv')
df_2016 = pd.read_csv('marathon_results_2016.csv')
df_2017 = pd.read_csv('marathon_results_2017.csv')

# Combine datasets
df=pd.concat([df_2015, df_2016, df_2017])

#Convert official Tine (HH:MM:SS) to seconds 
df[['Hours','Minutes','Seconds']] = df['Official Time'].str.split(':',expand = True)

df['Hours'] = df['Hours'].astype(int)
df['Minutes'] = df['Minutes'].astype(int)
df['Seconds'] = df['Seconds'].astype(int)

df['finish_time_in_seconds'] = (df['Hours'] *3600 + df['Minutes']* 60 + df['Seconds'])

#Take random sample of 30 runners
our_column = df['finish_time_in_seconds'].sample(n=30, replace = False)

#Null hypothesis mean = 3.75 hours 
meu = 3.75*3600

#Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(our_column, meu)

print("Sample Mean (Seconds): ", our_column.mean())
print("Sample Mean (hours): ", our_column.mean()/3600)
print("T-statistic: ", t_statistic)
print("P-value:", p_value)

alpha = 0.05

if p_value < alpha:
    
    print("Reject H0")
    
else:
    
    print("Do not reject H0")



```

# TWO SAMPLE T-TEST

#Two independent samples T-Test on Marathon datasets 

```python
# Do younger and older atheletes perform similarly?
```

```python
#Split the 'Pace' column into hours, minutes and seconds
#expand =True means split into different columns
df[['Hours', 'Minutes', 'Seconds']] = df['Pace'].str.split(':', expand=True)

#Convert hours, minutes, and seconds columns to integers

df['Hours'] = df['Hours'].astype(int)
df['Minutes'] = df['Minutes'].astype(int)
df['Seconds'] = df['Seconds'].astype(int)

#Manualy calculate total time in seconds

df['pace_in_seconds'] = (df['Hours'] *3600 + df['Minutes']* 60 + df['Seconds'])


#Separate data by Age
df['Age'] =df['Age'].astype(int)
young_athletes = df[(df['Age']<=30) & (df['M/F']=='M')]['pace_in_seconds']

older_athletes = df[(df['Age']>30) & (df['M/F']=='F')]['pace_in_seconds']

young_athletes = young_athletes.sample(n=30,replace=True)
older_athletes = older_athletes.sample(n=30, replace = True)
print(young_athletes.mean())
print(older_athletes.mean())

#Perform two-sample t-test

t_statistic, p_value = stats.ttest_ind(young_athletes, older_athletes)

#Degrees of freedom
n_young = len(young_athletes)
n_older = len(older_athletes)
degrees_of_freedom = n_young + n_older-2
#Significance level (Alpha)
alpha = 0.05
#Find the critical value
critical_value = stats.t.ppf(1-alpha,degrees_of_freedom)

#Print the results
print("T-statistic: ", t_statistic)
print("P-value: ", p_value)

if p_value < alpha:
    print("There is a significant difference")

else:
    print("There is no significant difference")




```

```python
#sample data
import pandas as pd
from scipy import stats

data = [501,502,503,499,500,504,501,502,500,503]

mu= 500

#--------------------------------------------------
#Two-tailed test
#--------------------------------------------------

result = stats.ttest_1samp(
    data,
    popmean = mu,
    alternative='two-sided'
)

print("\nTWO-TAILED TEST")
print("t-statistic =", round(result.statistic,4))
print("p-value  =",round(result.pvalue, 4))

#--------------------------------------------------
#Right-tailed test
#--------------------------------------------------
result = stats.ttest_1samp(
    data,
    popmean=mu,
    alternative='greater'
)

print("\nRIGHT-TAILED TEST")
print("t-statistic =", round(result.statistic,4))
print("p-value  =",round(result.pvalue, 4))

#--------------------------------------------------
#Left-Tailed TEST
#--------------------------------------------------

result=stats.ttest_1samp(
    data,
    popmean=mu,
    alternative='less'
)

print("\nLEFT-TAILED TEST")
print("t-statistic =", round(result.statistic,4))
print("p-value  =",round(result.pvalue, 4))
```

```python
#paired t-test: A company has launched some offer
#H0: Pre and post usage is the same (i.e offer is not effective)


from scipy.stats import ttest_rel
import pandas as pd

df=pd.read_csv("cust_seg.csv")
usage_data = df[["pre_usage","Post_usage_1month"]]
usage_data = usage_data.sample(n=30)

tstat,pval = ttest_rel(usage_data["pre_usage"],usage_data["Post_usage_1month"])

alpha = 0.05

if pval < alpha:
    print("reject null hypothesis")
else:
    print("Failed to reject null hypothesis")
```
