```python
#matplotlib and seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("tips.csv")
```

```python
#Line Plot: Average tip across row index
plt.figure(figsize=(6,4))
plt.plot(df.index,df["tip"],marker="+")
plt.title("Line Plot - Tip across records")
plt.xlabel("Record Index")
plt.ylabel("Tip")
plt.show()
```

```python
#2)Bar Plot : Average total_bill for each day
avg_bill_day = df.groupby("day")["total_bill"].mean()
plt.figure(figsize=(6,4))
plt.bar(avg_bill_day.index,avg_bill_day.values)
plt.title("bar plot - Average Total bill by day")
plt.xlabel("day")
plt.ylabel("average total bill")
plt.show()
```

```python
#3)Histogram : Dsitribution of Total_bill
plt.figure(figsize=(6,4))
plt.hist(df['total_bill'],bins=15,edgecolor = "black")
plt.title("Histogram - Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
#4. Box Plot: Tip spread by day
tip_by_day = [df[df['day']==d]["tip"] for d in df["day"].unique()]
plt.figure(figsize=(6,4))
plt.boxplot(tip_by_day, labels = df["day"].unique())
plt.title("Box Plot - Tip by day")
plt.xlabel("Day")
plt.ylabel("Tip")

```

```python
#4. Box Plot: Tip spread by day
tip_by_day = [df[df['day']==d]["tip"] for d in df["day"].unique()]
plt.figure(figsize=(6,4))
plt.boxplot(tip_by_day, labels = df["day"].unique())
plt.title("Box Plot - Tip by day")
plt.xlabel("Day")
plt.ylabel("Tip")

plt.show()
```

```python
#5. Scatter Plot: total_bill vs tip
plt.figure(figsize=(6,4))
plt.scatter(df["total_bill"],df["tip"])
plt.title("scatter plot - total bill v/s tip")
plt.xlabel("total bill")
plt.ylabel("tip")
plt.show()
```

```python
#6 Pie chart : Percentage of bills by day

day_counts=df["day"].value_counts()
plt.figure(figsize = (6,6))
plt.pie(day_counts.values,labels=day_counts.index,autopct="%1.1f%%")
plt.title("Piechart Chart - Bills By Day")
plt.show()
```

```python
#Seaborn

plt.figure(figsize=(6,4))
sns.lineplot(x=df.index,y=df["tip"])
plt.title("Seaborn Line Plot - Tip across records")
plt.xlabel("Record Index")
plt.ylabel("Tip")
plt.show()
```

```python
#2) Bar Plot
plt.figure(figsize = (6,4))
sns.barplot(x="day",y="total_bill",data=df)
plt.title("Seaborn Bar plot - Average total by Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill")
plt.show()

#The black lines that are appearing are the approximation of error ,it says actual bills vary above and below that average.
```

```python
#5)Scatter plot
plt.figure(figsize=(6,4))
sns.scatterplot(x="total_bill",y="tip",data=df)
plt.title("Seaborn Scatter plot - Total bill v/s tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```

```python
#6)Pie Chart
plt.figure(figsize=(6,6))
plt.pie(day_counts.values,labels=day_counts.index,autopct="%1.1f%%")
plt.title("Seaborn Section - Pie Chart by Day")
plt.show()
```

```python
#7) Heatmap
corr = df[["total_bill","tip","size"]].corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Seaborn Heatmap - CorrelationMatrix")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv("tips.csv")
avg_tip_by_day=df.groupby("day")["tip"].mean()
corr=df[["total_bill","tip","size"]].corr()

#Functional syntax for subplots

plt.figure(figsize=(12,8))

#Line
plt.subplot(2,2,1)
plt.plot(df.index,df["total_bill"])
plt.title("Line plot")

#Bar
plt.subplot(2,2,2)
plt.bar(avg_tip_by_day.index,avg_tip_by_day.values)
plt.title("Bar Plot")

#Histogram
plt.subplot(2,2,3)
plt.hist(df["total_bill"],bins=10)
plt.title("Histogram")

#Box
plt.subplot(2,2,4)
plt.boxplot(df["tip"])
plt.title("Box plot")

plt.tight_layout() #If there is any distortion in the graph, tight_layout will try to adjust the size,
#we usually use it in the subplots, there is no gaurantee it will fix if we don't give the correct size
plt.show()
```

```python
#object oriented style

fig,axes = plt.subplots(2,2,figsize=(12,8))

#Line
axes[0,0].plot(df.index,df["total_bill"])
axes[0,0].set_title("Line plot")

#Bar
axes[0,1].bar(avg_tip_by_day.index,avg_tip_by_day.values)
axes[0,1].set_title("bar plot")

#Box
axes[1,0].boxplot(df["tip"])
axes[1,0].set_title("Box plot")

#Scatter
axes[1,1].scatter(df["total_bill"],df["tip"])
axes[1,1].set_title("scatter plot")

plt.tight_layout()
plt.show()
```
