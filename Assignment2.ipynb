```python
#Q1 Do an outlier analysis by displaying q1,q2,q3,IQR,LB,UB, Outlier points and also using box plot on the data set
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("Billionaires Statistics Dataset.csv")
final_net_worth = np.array(df["finalWorth"])
```

```python
#outlier analysis
q1 = np.percentile(final_net_worth,25)
q3 = np.percentile(final_net_worth,75)

iqr = q3-q1
lb = q1 - (1.5 * iqr)
ub = q3 - (1.5 * iqr)

print(lb)
print(ub)
print(iqr)
```

```python
df
```

```python
outliers = []
for x in final_net_worth:
    if x < lb or x>lb:
        outliers.append(x)
print(outliers)
```

```python
plt.figure(figsize = (10,6))
sns.boxplot(data=final_net_worth,orient = 'h')
plt.title("Billinaires outlier analysis using box plot")
plt.xlabel("finalWorth")
plt.show()
```

```python
#Q3 Use mall-customers.csv dataset
#Perform all the four the sampling techniques to pick 10 % records
#For stratified random sampling,(a) first consider gender (b) Then consider anual income (slabs can be <= 50, 51-99, 100-149)
#For cluster random sampling, Consider age brackets (19-29,30-39,....)
#Find average spending score for each of the sampled subset (x bar) 
#Which one is closest to population range(meu)?


#Simple sampling

import pandas as pd
import numpy as np

df = pd.read_csv("Mall_Customers.csv")

original_row_count = len(df)

#Perform simple random sampling (selecting 1/10th of the records)

sample_df = df.sample(frac=0.1,random_state=42)

#Get the total number of rows in the sample dataframe 
sample_row_count = len(sample_df)

print(df.head())
print(sample_df.head())

#Print the number of rows in the original sample data frame

print(f"Number of rows in the original dataframe {original_row_count}")
print(f"Number of rows in the sample dataframe: {sample_row_count}")


```

```python
#systematic sampling
#Get the total number of rows in the original dataframe

original_row_count = len(df)

#Perform systematic sampling (Picking every 7th record)
#::7 means from the start to the end, pick evry 7th record
#, : means pickup all the columns
sampled_df = df.iloc[::7, :]

#Get the total number of rows in the sampled dataframe
sampled_row_count = len(sample_df)

#Print first few rows of the original and sampled dataframe
print("first few rows of the original dataframe")
print(df.head())
print("\n first few rows of the original dataframe")

print(sampled_df.head())
print(sampled_df.tail())
```

```python
import pandas as pd


population_mean = df['Spending Score (1-100)'].mean()
print("Population mean:", population_mean)


Age_1 = df[df['Age'].between(19,29)]
Age_2 = df[df['Age'].between(30,39)]
Age_3 = df[df['Age'].between(40,59)]
Age_4 = df[df['Age'].between(60,69)]
Age_5 = df[df['Age'].between(70,79)]
Age_6 = df[df['Age'].between(80,89)]
Age_7 = df[df['Age'].between(90,99)]


age_brackets = [Age_1, Age_2, Age_3, Age_4, Age_5, Age_6, Age_7]

# Compute average spending score (x̄) for each cluster
cluster_means = []
for i, cluster in enumerate(age_brackets, start=1):
    mean_score = cluster['Spending Score (1-100)'].mean()
    cluster_means.append((i, mean_score))
    print(f"age_bracket {i} mean spending score (x̄): {mean_score:.2f}")

closest_cluster = min(cluster_means, key=lambda x: abs(x[1] - population_mean))
print(f"\nage_bracket {closest_cluster[0]} is closest to population mean  ({population_mean:.2f})")
```
