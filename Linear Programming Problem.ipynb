```python
#Linear Programming Problem(LPP)
# pip install pulp
from pulp import *

#Create the variable to contain the problem data
model =  LpProblem("FactoryProductionProblem", LpMaximize)

#Create variables for Chairs and Tables
#Parameters : Name,Lower limit, Upper limit, Data type

chairs = LpVariable("Chairs",0,None,LpInteger) #Number of chairs to produce
tables = LpVariable("Tables",0,None,LpInteger)#Number of tables to produce

#create maximize objective function(profit)
model+=300 * chairs +1000 * tables, "Profit"

#create constraints
model+=5 * chairs + 10 * tables <= 400 , "wood Constraint" #Wood availability constraint
model+=2 * chairs + 6 * tables <=300, "Time Constraint" #Worker time constraint

#The problem is solved using PuLP's choice of Solver
model.solve()
for v in model.variables():
    print(v.name, "=", v.varValue)

#Additional Information
print("Total Profit:", value(model.objective)) #Print objective function value ( Total profit)
```

```python
#Linear Programming Problem(LPP)
# pip install pulp
from pulp import *

#Create the variable to contain the problem data
model =  LpProblem("BankLoanProblem", LpMinimize)

#Create variables for CarA and CarB
#Parameters : Name,Lower limit, Upper limit, Data type

BankA = LpVariable("Bank A",0,None,LpInteger) #Bank A laon
BankB = LpVariable("Bank B",0,None,LpInteger) #Bank B loan
BankC = LpVariable("Bank C",0,None,LpInteger) #Bank C loan
#create maximize objective function(profit)

model+=2076 * BankA+ 3227 * BankB + 4707 * BankC, "TotalMonthlyEMI"


#create constraints
model+=BankA >= 2,"lower risk"
model+=BankC <= 4,"higher interest"
model+= BankA+BankB+BankC == 10,"Total loan"


#The problem is solved using PuLP's choice of Solver
model.solve()
for v in model.variables():
    print(v.name, "=", v.varValue)

#Additional Information
print("Total EMI:", value(model.objective)) #Print objective function value ( Total profit)

```
