```python

print("==================creating array=====================")
arr = np.array([1,2,3,4])
lst = [1,2,3,4]
print("==================Add  5=====================")
print(arr+5)
print([x+5 for x in lst])
print("==================multiply by 2=====================")
print(arr*2)
print([x*2 for x in lst])
print("==================Squares=====================")
print(arr**2)
print([x**2 for x in lst])
print("==================Mean=====================")
print(np.mean(arr))
print(sum(lst)/len(lst))
print("==================Filtering factors of 2=====================")
print(arr[arr % 2 == 0])
print([x for x in lst if x%2==0])
print("==================2D Array=====================")
print(np.array([[1,2],[3,4]]))
print([[1,2],[3,4]])
print("==================Adding Two Arrays=====================")
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a+b)

a=[1,2,3]
b=[4,5,6]
print([x+y for x,y in zip(a,b)])

print("==================Broadcasting=====================")
'''Broadcasting is a mechanism that allows arrays of different shapes to perform arithmetic operations together without 
explicitly resizing or copying data'''
print(arr+1)
#python list: Not directly supported like NumPy

print("===================Condtional Replace===============")
#Conditional Replace
arr[arr>2] = 0
print(arr)

lst = [1,2,3,4]
print([0 if x>2 else x for x in lst])
```

```python
#shallow v/s deep copy
a=np.array([10,20,30,40,50,60])

#shallow copy
b = a[0:5]
print(a)
print(b)

b[0] = 999
print(a)
print(b)

#Deep Copy
c=a.copy()

c[1] = 888
print(a)
print(c)

```

```python
#loc and iloc
import pandas as pd

data={
    "Player":['Virat','Rohit','MS Dhoni'],
    "Match1":[45,55,30],
    "Match2":[67,72,40],
    "Match3":[82,60,50],
    "Match4":[82,60,50],
    "Match5":[82,60,50],
    "Match6":[82,60,50],
    "Match7":[82,60,50]
}

df = pd.DataFrame(data).set_index("Player")
print(df)

#First row
print("------------------------------First row-------------------------")

print(df.loc["Virat"].values)
print(df.iloc[0].values)

#print(df.head())

#subset
print("------------------------------Subset-------------------------")

print(df.loc[["Virat","MS Dhoni"],["Match1"]])

#slicing
print("------------------------------Slicing-------------------------")

# From Virat to Rohit (inclusive)
print(df.loc["Virat":"Rohit", "Match1":"Match3"])
# First two rows, first three columns
print(df.iloc[0:2, 0:3])

df = pd.DataFrame({"Employee" : ["Vachan","Sreenath","Nikhil", "Aashutosh","Krutik","Rohit"],
                    "Department":["IT","HR","IT","HR","Fin","FIn"],
                    "Salary": [50000,np.nan,100000,450000,50000,74000]})
print(df)

#INDEXING
print("-----------------------INDEXING----------------------")
df_indexed = df.set_index("Employee")
print(df_indexed)

#Filtering
print("-----------------Filtering-----------------")

print(df[df["Salary"] > 50000])

#Check missing
print("------------check missing------------------")

print(df.isnull())

#Fill missing
print("-----------------Fill missing-----------")
print(df.fillna(220))
print("-----------------Drop na-----------")

#Drop missing
print(df.dropna())
print("-----------------Grouping-----------")

#grouping
print(df.groupby("Department")["Salary"].mean())

print("-----------------merge-----------")

#Merge (Join)
df_bonus = pd.DataFrame({
    "Employee":["Rohit","Nikhil","Raj"],
    "Bonus":[5000,7000,4000]
})
merged = pd.merge(df, df_bonus, on="Employee")
print(merged)


merged  = pd.merge(df,df_bonus,on="Employee",how="left")
print(merged)

print("-----------------Concat-----------")

#concat

df_new = pd.DataFrame({
    "Employee":["Kiran"],
    "Department":["IT"],
    "Salary":[55000]
})

concat_df = pd.concat([df,df_new], ignore_index=True)
print(concat_df)
```

```python
import pandas as pd
df = pd.read_csv("tips.csv")
print(df)
print(df.head())
print(df[df['total_bill']>20].head())
print(df[(df['total_bill']>20) & (df['tip'] < 3 )].head())
print(df.loc[df['size']==2, ['total_bill','tip']].head())

print(df.query("total_bill >20 and tip <3").head())
print(df[df['day'].isin(['Sat', 'Sun'])].head())
print(df[df['total_bill'].between(15,25)].head())
print(df[df['gender'].astype(str).str.contains("Male")].head())
```

```python
# EDA (EXPLORATORY DATA ANALYSIS

df =pd.read_csv("tips.csv")
print(df.head())
print(df.tail())
print(df.shape)
print(df.columns)
print(df.dtypes)
print(df.info())
print(df.isnull().sum())
print(df.describe())
print(df['day'].unique())

#Total bill analysis

print(df['total_bill'].min())

print(df['total_bill'].max())

print(df['total_bill'].mean())
print(df['total_bill'].median())

print(df.groupby('day')['total_bill'].mean())
print(df.groupby('gender')['tip'].mean())

print(df.nlargest(5,'total_bill'))
print(df.nsmallest(5,'total_bill'))

print(df[df['total_bill']>30].head())
print(df[df['smoker'] == 'Yes'].head())

print(df[df['day'].isin(['Sat','Sun'])].head())

df['Tip_Percentage'] = (df['tip']/df['total_bill'])*100
print(df)

print("----------------------numpy operation--------------------------------")
#numpy operations
bill_array = np.array(df['total_bill'])
print(round(np.mean(bill_array),2))
print(round(np.median(bill_array),2))

print(np.std(bill_array))

print(df['size'].value_counts())
print(df['smoker'].value_counts())


```
