## Simple random sampling

```python

import pandas as pd

df_2015 = pd.read_csv("marathon_results_2015.csv")
df_2016 = pd.read_csv("marathon_results_2016.csv")
df_2017 = pd.read_csv("marathon_results_2017.csv")

#https://github.com/newdelthis/demo/


#Concatenate the datasets into a single dataframe
df = pd.concat([df_2015,df_2016,df_2017])

#Get the total number of rows in the original dataframe

original_row_count = len(df)

#Perform simple random sampling (selecting 1/10th of the records)

sample_df = df.sample(frac=0.1,random_state=42)

#Get the total number of rows in the sample dataframe 
sample_row_count = len(sample_df)

print(df.head())
print(sample_df.head())

#Print the number of rows in the original sample data frame

print(f"Number of rows in the original dataframe {original_row_count}")
print(f"Number of rows in the sample dataframe: {sample_row_count}")
```

## Systematic random sampling

```python
#Get the total number of rows in the original dataframe

original_row_count = len(df)

#PErform systematic sampling (Picking every 7th record)
#::7 means from the start to the end, pick evry 7th record
#, : means pickup all the columns
sampled_df = df.iloc[::7, :]

#Get the total number of rows in the sampled dataframe
sampled_row_count = len(sample_df)

#Print first few rows of the original and sampled dataframe
print("first few rows of the original dataframe")
print(df.head())
print("\n first few rows of the original dataframe")

print(sampled_df.head())
```

## Stratified random sampling

```python
#Get the total number of rows in the original dataframe
original_row_count = len(df)

#Find the total number of male and female athletes

toatl_males= len(df[df['M/F']=="M"])

toatl_females= len(df[df['M/F']=="F"])

#Calculate proportionate number of male and female atheletes for sample

sample_size=int(0.1*original_row_count)#taking 
sample_males = int(0.1*toatl_males)
sample_females = int(0.1*toatl_females)

print(f"sample_size:{sample_size}\nsample_males:{sample_males}\nsample_females:{sample_females}")
```

```python
#Generate proportionate samples of male and female atheletes

male_sample = df[df['M/F']=="M"].sample(n=sample_males,random_state=42)
female_sample = df[df['M/F']=="F"].sample(n=sample_females,random_state=42)
print(len(male_sample),len(female_sample))

#combine male and female samples into single sample dataframe
sample_df = pd.concat([male_sample,female_sample])
#get the total number of rows in the sample dataframe
sampled_row_count = len(sample_df)


#print first few rows of the original and sampled data frames
print("first few rows of the original data frame")
print(df.head())
print("first few rows of the sampled data frame")
print(sample_df.head())
#print number of rows in original and sampled dataframes 
print("number of rows in original")
original_row_count = len(df)
print(original_row_count)

print("number of rows in sampled dataframe")
print(sample_row_count)
```

## Cluster sampling

```python
#Cluster sampling
#Find the unique countries the athletes are from
import numpy as np
unique_countries = df['Country'].unique()
num_countries = len(unique_countries)

print(f"unique_countries:\n{unique_countries}")
print(f"num_countries:{num_countries}")

# randomly select half of the countries
np.random.seed(101)

# choice() function in Numpy geneartes a random sample from an array
selected_countries = np.random.choice(unique_countries,size = num_countries//2,replace=False)

#replace = False means don't allow the same country twice or more times
print(f"selected_countries:\n{selected_countries}")

#create a new dataframe containing all records of the selected countries
sample_df = df[df["Country"].isin(selected_countries)]

#get the total number of rows in sample dataframe
sample_row_count = len(sample_df)

print("sample_row_count:{sample_row_count}")

#get the first few rows of original and sampled data

print("\nFirst few rows of original dataframe:")
print(df.head())

print("\nFirst few rows of sampled dataframe:")
print(sample_df.head())
```
