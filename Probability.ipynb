## probabilities : Marginal , Joint (Independent and Dependent Events)

```python
import pandas as pd

df = pd.read_csv("titanic.csv")
selected = ['gender','survived']
df = df[selected]
df.dropna(inplace=True)

Total = len(df)
print(f"Total passengers: {Total}")

# Marginal probability of gender
Total_female = len(df[df['gender']=='female'])
Total_male = len(df[df['gender']=='male'])

p_female = Total_female / Total
p_male = Total_male / Total

# Marginal probability of survival
Total_survived = len(df[df['survived']==1])
Total_didnot_survived = len(df[df['survived']==0])

p_survived = Total_survived / Total
p_not_survived = Total_didnot_survived / Total

print("\nMarginal probabilities:")
print(f"P(gender=female): {p_female:.3f}")
print(f"P(gender=male): {p_male:.3f}")
print(f"P(survived=1): {p_survived:.3f}")
print(f"P(survived=0): {p_not_survived:.3f}")

# Joint probabilities
p_female_survived = len(df[(df['gender']=='female') & (df['survived']==1)]) / Total
p_female_not_survived = len(df[(df['gender']=='female') & (df['survived']==0)]) / Total
p_male_survived = len(df[(df['gender']=='male') & (df['survived']==1)]) / Total
p_male_not_survived = len(df[(df['gender']=='male') & (df['survived']==0)]) / Total

print("\nJoint probabilities:")
print(f"P(female, survived=1): {p_female_survived:.3f}")
print(f"P(female, survived=0): {p_female_not_survived:.3f}")
print(f"P(male, survived=1): {p_male_survived:.3f}")
print(f"P(male, survived=0): {p_male_not_survived:.3f}")

# Conditional probabilities
p_survived_given_female = p_female_survived / p_female
p_survived_given_male = p_male_survived / p_male

print("\nConditional probabilities:")
print(f"P(survived | female): {p_survived_given_female:.3f}")
print(f"P(survived | male): {p_survived_given_male:.3f}")
```
